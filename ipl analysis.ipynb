## 2. Batting Analysis

#### Data Analysis

- Top 10 run scorers

```python
batter_summary = df.groupby('batter').agg({
    'runs_batter': 'sum'
}).reset_index()
top_10_batters = batter_summary.sort_values('runs_batter', ascending=False).head(10)
top_10_batters
```

- Top 10 strike rates (min 100 balls)

```python
batter_summary = df.groupby('batter').agg({
    'runs_batter': 'sum',
    'balls_faced': 'sum'
}).reset_index()
filtered_batters = batter_summary[batter_summary['balls_faced'] > 100].copy()
filtered_batters['strike_rate'] = (filtered_batters['runs_batter'] / filtered_batters['balls_faced']) * 100
top_10_strike_rate = filtered_batters.sort_values('strike_rate', ascending=False).head(10)
top_10_strike_rate[['batter', 'runs_batter', 'balls_faced', 'strike_rate']]
```

- Top 10 consistent batters (min 30 matches)


```python

batter_summary = df.groupby('batter').agg({
    'runs_batter': 'sum',
    'balls_faced': 'sum',
    'bat_pos': 'mean',
    'match_id': pd.Series.nunique  # number of unique matches per batter
}).rename(columns={'match_id': 'matches'}).reset_index()
consistent_batters = batter_summary[batter_summary['matches'] > 30]
top_10_consistent = consistent_batters.sort_values('runs_batter', ascending=False).head(10)
top_10_consistent[['batter', 'matches', 'runs_batter']]
```

- Average Batting Position of Top Run Scorers

```python
top_run_scorers = batter_summary.sort_values('runs_batter', ascending=False)[['batter', 'runs_batter', 'bat_pos']].head(10)
top_batting_pos = top_run_scorers[['batter', 'bat_pos']]
top_batting_pos
```

- Batters with Most Matches Played


```python
most_matches = batter_summary.sort_values('matches', ascending=False).head(10)
most_matches[['batter', 'matches']]
```

- Most Runs in a Single Match by a Batter

```python
runs_per_match = df.groupby(['match_id', 'batter'])['runs_batter'].sum().reset_index()
top_match_scores = runs_per_match.sort_values('runs_batter', ascending=False).head(10)
top_match_scores
```

- Batters Who Faced the Most Balls

```python
batter_summary = df.groupby('batter').agg({
    'runs_batter': 'sum',
    'balls_faced': 'sum',
    'bat_pos': 'mean',
    'match_id': pd.Series.nunique
}).rename(columns={'match_id': 'matches'}).reset_index()
most_balls = batter_summary.sort_values('balls_faced', ascending=False).head(10)
most_balls[['batter', 'balls_faced']]
```

- Best Ball-to-Run Ratio (min 500 balls faced)

```python
efficiency = batter_summary[batter_summary['balls_faced'] > 500].copy()
efficiency['ball_run_ratio'] = efficiency['balls_faced'] / efficiency['runs_batter']
best_efficiency = efficiency.sort_values('ball_run_ratio').head(10)
best_efficiency[['batter', 'ball_run_ratio']]
```

#### Visualization

- Top 10 run scorers

```python
# Group by batter and calculate total runs
batter_summary = df.groupby('batter').agg({
    'runs_batter': 'sum',
    'balls_faced': 'sum',
    'bat_pos': 'mean',
    'match_id': pd.Series.nunique
}).rename(columns={'match_id': 'matches'}).reset_index()

# Get top 10 run scorers
top_batters = batter_summary.sort_values('runs_batter', ascending=False).head(10)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=top_batters, x='runs_batter', y='batter', palette='ocean')
plt.title('Top 10 Run Scorers in IPL', fontsize=16, weight='bold')
plt.xlabel('Total Runs Scored', fontsize=12)
plt.ylabel('Batter Name', fontsize=12)


plt.tight_layout()
```

- Top 10 Strike Rates (min 100 balls)

```python
# Group by batter and calculate total runs and balls faced
batter_summary = df.groupby('batter').agg({
    'runs_batter': 'sum',
    'balls_faced': 'sum'
}).reset_index()

# Filter batters who faced more than 100 balls
filtered_batters = batter_summary[batter_summary['balls_faced'] > 100].copy()

# Calculate strike rate
filtered_batters['strike_rate'] = (filtered_batters['runs_batter'] / filtered_batters['balls_faced']) * 100

# Top 10 strike rates
top_strike_rates = filtered_batters.sort_values('strike_rate', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_strike_rates, x='strike_rate', y='batter', palette='magma')
plt.title('Top 10 Batters by Strike Rate (min 100 balls)', fontsize=16)
plt.xlabel('Strike Rate', fontsize=12)
plt.ylabel('Batter', fontsize=12)
plt.tight_layout()
plt.show()
```

- Top Consistent Batters (min 30 matches)

```python
# Create batter summary if not already done
batter_summary = df.groupby('batter').agg({
    'runs_batter': 'sum',
    'balls_faced': 'sum',
    'bat_pos': 'mean',
    'match_id': pd.Series.nunique  # unique matches
}).rename(columns={'match_id': 'matches'}).reset_index()

# Filter batters who played more than 30 matches
consistent_batters = batter_summary[batter_summary['matches'] > 30]

# Get top 10 by total runs
top_consistent = consistent_batters.sort_values('runs_batter', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_consistent, x='runs_batter', y='batter', palette='inferno')
plt.title('Top Consistent Batters (30+ matches)', fontsize=16)
plt.xlabel('Runs Scored', fontsize=12)
plt.ylabel('Batter', fontsize=12)
plt.tight_layout()
```

### 4. Team-wise & All-Rounder Insights

- Team-wise Total Wickets

```python
team_wickets = df[df['bowler_wicket'] > 0].groupby('bowling_team')['bowler_wicket'].sum().reset_index()
team_wickets
```

- Team with Best Bowling Average (runs/wicket)

```python
team_runs = df.groupby('bowling_team')['team_runs'].sum()
team_avg = pd.merge(team_wickets, team_runs, on='bowling_team')
team_avg['avg_per_wicket'] = team_avg['team_runs'] / team_avg['bowler_wicket']
team_runs
team_avg
```

- Top All-rounders: 30+ wickets and 500+ runs

```python
# Step 1: Batter stats
batter_stats = df.groupby('batter')['batter_runs'].sum().reset_index()
# Step 2: Merge with wicket counts
all_rounders = pd.merge(batter_stats, wicket_counts, left_on='batter', right_on='bowler')
top_all_rounders = all_rounders[(all_rounders['batter_runs'] > 500) & (all_rounders['wickets'] > 30)]
top_all_rounders
```

```python
all_rounders
```

- Top Run Scorer Among All-Rounders

```python
top_all_rounders.sort_values('batter_runs', ascending=False).head(1)
```

- Most 4-Wicket Hauls

```python
hauls = df[df['bowler_wicket'] > 0].groupby(['match_id', 'bowler'])['bowler_wicket'].sum()
haul_counts = hauls[hauls >= 4].groupby('bowler').count().reset_index(name='4_wicket_hauls')
haul_counts
```

- Bowler with Highest Dot Ball %

```python
dots = df[(df['team_runs'] == 0) & (df['bowler_wicket'] == 0)]
dot_balls = dots.groupby('bowler').size().reset_index(name='dot_balls')
dot_balls  


```

- Most Common Bowling Team

```python
df['bowling_team'].value_counts().reset_index().head(1)
```

- Bowler with Most Matches



```python
matches_played = df.groupby('bowler')['match_id'].nunique().reset_index(name='matches')
matches_played.sort_values('matches', ascending=False).head(1)
```

- Team-wise Bowling Strike Rate

```python
team_balls = df.groupby('bowling_team').size().reset_index(name='balls')
team_strike = pd.merge(team_wickets, team_balls, on='bowling_team')
team_strike['team_strike_rate'] = team_strike['balls'] / team_strike['bowler_wicket']
team_balls
team_strike
```
